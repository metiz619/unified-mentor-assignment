### Create colors = [‘red’,’blue’,’orange’], and fruits = [‘orange’,’grapes’,’apples’]

```python
colors=['red','blue','orange']
fruits=['orange','grapes','apples']
```

Print color and fruits

```python
print(colors)
print(fruits)
```

Create colors_set, and fruits_set. (using set() ) and print them

```python
colors_set=set(colors)
fruits_set=set(fruits)
print(colors_set)
print(fruits_set)
```

Find the union of both the sets.

```python
union_set_method=colors_set.union(fruits_set)
print(union_set_method)

```

Find the intersection of both the sets

```python
intersection_set_method=colors_set.intersection(fruits_set)
print(intersection_set_method)

```

Find the elements which are Fruits but not colors (using set.difference() )

```python
fruits_not_in_colors=fruits_set.difference(colors_set)
print(fruits_not_in_colors)
```

count the number of occurences of charachters in string named "sent" using dictionary and print the same.


```python
char_counts={}

```

```python
for char in sent:
    if char.isalpha():  # Ignore non-alphabetic characters
        char = char.lower()  # Consider case-insensitive counting
        char_counts[char] = char_counts.get(char, 0) + 1
```

```python
for char,count in char_counts.items():
    print(f"{char} : {count}")
```
